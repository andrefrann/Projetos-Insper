```python
def DeltaA(A,an,am,AV,V,TA,T):
    DA = (an-am)*A + VA*V*A - TA*A*T
    return DA

def DeltaV(V,vn,vm,AV,A):
    DV = (vn-vm)*V - AV*V*A
    return DV

def DeltaT(T,tn,tm,AT,A,ct):
    DT = (tn-tm)*T - ct*T + AT*A*T
    return DT

tempo = int(input("Qual o tempo da previsão?"))

A = [400]
an = 0.07
am = 0.09
VA = 0.0001
TA = 0.0005

V = [1500]
vn = 0.3
vm = 0.2
AV = 0.0006

T = [20]
tn = 0.001
tm = 0.005
AT = 0.0002
ct = 0.079

for i in range(tempo):    
    DA = DeltaA(A[i], an, am, AV, V[i], TA, T[i])
    A_novo = A[i] + DA
    A.append(A_novo)
    
    DV = DeltaV(V[i], vn, vm, AV, A[i])
    V_novo = V[i] + DV
    V.append(V_novo)
    
    DT = DeltaT(T[i], tn, tm, AT, A[i], ct)
    T_novo = T[i] + DT
    T.append(T_novo)


import matplotlib.pyplot as plt 
plt.plot(V)
plt.plot(V, linestyle="-", color = "blue", label=u"Vieiras")
plt.xlabel("Tempo") 

plt.plot(A)
plt.plot(A, linestyle="-", color = "green", label=u"Arraias")
 
plt.plot(T)
plt.plot(T, linestyle="-", color = "red", label=u"Tubarões")
plt.grid()
plt.legend(loc="upper right")
plt.title("Populações")
plt.show()
```
